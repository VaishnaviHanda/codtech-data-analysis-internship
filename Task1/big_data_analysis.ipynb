```python
import dask.dataframe as dd

df = dd.read_csv("netflix.csv", assume_missing=True)

df.head()
```

```python
df.dtypes
```

```python
df_clean = df.dropna()

print("Rows Before:", len(df))
print("Rows After:", len(df_clean))
```

```python
print("Movies vs TV Shows:")
print(df_clean['type'].value_counts().compute())

print("\nTop 10 Countries:")
print(df_clean['country'].value_counts().head(10))

print("\nContent by Release Year:")

year_counts = df_clean['release_year'].value_counts().compute()
year_counts = year_counts.sort_index()

print(year_counts)
```

```python
import matplotlib.pyplot as plt


type_counts = df_clean['type'].value_counts().compute()

type_counts.plot(kind='bar')

plt.title("Movies vs TV Shows on Netflix")
plt.xlabel("Type")
plt.ylabel("Count")
plt.grid(True)

plt.show()
```
